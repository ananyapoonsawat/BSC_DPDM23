#ประวัติการรับทุนของบุคลากร

##นำเข้าข้อมูล ประวัติการรับทุนของบุคลากร

```python
dv = pd.read_excel('/content/drive/MyDrive/ข้อมูล กองทรัพย์/ประวัติการรับทุนของบุคลากร.xlsx') #เรียกใช้ไฟล์ที่ต้องการ และ ตั้งหัวตาราง
dv
```

##เปลี่ยนข้อมูลคอลัมล์ columns ประวัติการรับทุนของบุคลากร

```python
dv['course'] = dv['course'].str.replace('ปริญญาเอกร่วมกับมหาวิยาลัยในต่างประเทศ','ปริญญาเอกร่วมกับมหาวิทยาลัยในต่างประเทศ')
dv['course'] = dv['course'].str.replace('ปริญญาเอกร่วมร่วมกับมหาวิทยาลัยในต่างประเทศ','ปริญญาเอกร่วมกับมหาวิทยาลัยในต่างประเทศ')
```

```python
dv['major'] = dv['major'].str.replace('Nursitng','Nursing')
dv['major'] = dv['major'].str.replace('Romete Sensing & GIS','Remote Sensing & GIS')
dv['major'] = dv['major'].str.replace('Remote Sensing and GIS','Remote Sensing & GIS')
dv['major'] = dv['major'].str.replace('เทคโนโลยีการอาหาร','เทคโนโลยีทางอาหาร')
dv['major'] = dv['major'].str.replace('เทคโนโลยีอาหาร','เทคโนโลยีทางอาหาร')
```

```python
dv['place'] = dv['place'].str.replace('Linclon University','Lincoln University')
dv['place'] = dv['place'].str.replace('มหาวิทยาลัยศรีนครินทร์วิโรฒ','มหาวิทยาลัยศรีนครินทรวิโรฒ')
dv['place'] = dv['place'].str.replace('มหาวิทยาลัยเกตษรศาสตร์','มหาวิทยาลัยเกษตรศาสตร์')
dv['place'] = dv['place'].str.replace('คณะแพทยศาสตร์ ม.ขอนแก่น','คณะแพทยศาสตร์ มหาวิทยาลัยขอนแก่น')
dv['place'] = dv['place'].str.replace('มหาวิทยาลัยขอนแกน','มหาวิทยาลัยขอนแก่น')
dv['place'] = dv['place'].str.replace('คณะพยาบาลศาสตร์ ม.ขอนแก่น','คณะพยาบาลศาสตร์ มหาวิทยาลัยขอนแก่น')
dv['place'] = dv['place'].str.replace('คณะแพทยศาสตร์ ม.มหิดล','คณะแพทยศาสตร์ มหาวิทยาลัยมหิดล')
dv['place'] = dv['place'].str.replace('มหาหาวิทยาลัยขอนแก่น','มหาวิทยาลัยขอนแก่น')
dv['place'] = dv['place'].str.replace('มหาวิทยาัยขอนแก่น','มหาวิทยาลัยขอนแก่น')
dv['place'] = dv['place'].str.replace('สถาบันบัณฑิตพัฒนาบริหารศาสตร์','สถาบันบัณฑิตพัฒนบริหารศาสตร์')
dv['place'] = dv['place'].str.replace('รพ.จุฬาลงกรณ์','โรงพยาบาลจุฬาลงกรณ์')
dv['place'] = dv['place'].str.replace('สถาบันเทคโนโลยีพระจอมเกล้าลาดกระบัง','สถาบันเทคโนโลยีพระจอมเกล้าเจ้าคุณทหารลาดกระบัง')
```

```python
dv['country'] = dv['country'].str.replace('ไทบ','ไทย')
dv['country'] = dv['country'].str.replace('เยอรมัน','เยอรมนี')
dv['country'] = dv['country'].str.replace('สหหรัฐอเมริกา','สหรัฐอเมริกา')
```
