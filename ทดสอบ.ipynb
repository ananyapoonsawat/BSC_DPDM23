#เปลี่ยนข้อมูลคอลัมล์ columns ประวัติการให้ทุนการศึกษาบุตรบุคลากร

```python
data['name'] = data['name'].str.replace('เด็ฏชาย','ด.ช.')
data['name'] = data['name'].str.replace('เด็กชาย','ด.ช.')
data['name'] = data['name'].str.replace('เด็กหญฺง ','ด.ญ.')
data['name'] = data['name'].str.replace('เด้กหญิง','ด.ญ.')
data['name'] = data['name'].str.replace('เด็กหญิง','ด.ญ.')

data['name'] = data['name'].str.replace('ดช.','ด.ช.')
data['name'] = data['name'].str.replace('ดญ.','ด.ญ.')

data['name'] = data['name'].str.replace('ด.ช. ','ด.ช.')
data['name'] = data['name'].str.replace('ด.ญ. ','ด.ญ.')

data['name'] = data['name'].str.replace('ดช. ','ด.ช.')
data['name'] = data['name'].str.replace('ดญ. ','ด.ญ.')

data['name'] = data['name'].str.replace('นาย ','นาย')

data['name'] = data['name'].str.replace('นาง ','นาง')

data['name'] = data['name'].str.replace('นางสาว','น.ส.')
data['name'] = data['name'].str.replace('นางสาว ','น.ส.')
data['name'] = data['name'].str.replace('นางสาง','น.ส.')

data['name'] = data['name'].str.replace('น.ส. ','น.ส.')
data['name'] = data['name'].str.replace('นส.','น.ส.')

data['name']
```
