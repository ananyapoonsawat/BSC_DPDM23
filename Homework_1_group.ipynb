```python
data_ChristmasNewYear = data_UNset2[data_UNset2['month']==12] #data_UNset2คือเซทที่เคลียร์ล่สุดแล้ว
data_ChristmasNewYear
```

```python
transactionsChristmasNewYear = data_ChristmasNewYear.groupby('InvoiceNo')['StockCode'].apply(list)
```

```python
frequentItemsets = list(apriori(transactionsChristmasNewYear,min_support = 0.05))
#frequentItemsets
```
